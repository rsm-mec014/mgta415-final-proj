# Preprocessing

### Tokenization

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer() 

stem_doc = []
for sentence in df['review'].values:
    tok = word_tokenize(sentence)
    words = [ps.stem(word) for word in tok]
    stem_doc.append(words)

stem_doc[:5]
```
